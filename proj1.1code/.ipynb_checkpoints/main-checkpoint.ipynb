## Logic Based FizzBuzz Function [Software 1.0]

```python
import pandas as pd #imports pandas methods as pd since python built-in methods can overlap panda methods. 

def fizzbuzz(n): #define function Fizzbuzz.
    
    # Logic Explanation : 
    # The logic is simply based on the working of modulus fuction. 
    # Say we have a number which is dividibe by 3, e.g. 9 / 3 = 3.
    # Since 9 is completely divisible by 3, the modulo returns 0(Fizz) as output else returns the remainder.
    # Say for 10 / 3 = 1, the modulo will return 1 as the output.
    # Similarly for a number divisible by 5 output = 0(Buzz) else the remainder.
    # Also for the numbers divisible by 3 & 5 respectivly the modulo will return 0(FizzBuzz) else remainder.
    
    if n % 3 == 0 and n % 5 == 0: #if output of modulo for 3 & 5 = 0
        return 'FizzBuzz' #return FizzBuzz
    elif n % 3 == 0: #if output of modulo for 3 = 0
        return 'Fizz' #return Fizz
    elif n % 5 == 0: #if output of modulo for 5 = 0
        return 'Buzz' #return Buzz
    else:
        return 'Other' #return Other if not completely divisible by 3 or 5 or 15 
```

## Create Training and Testing Datasets in CSV Format

```python
def createInputCSV(start,end,filename): #define tabular file format of the Input file with parameters required.
    
    # Why list in Python? : 
    # As python Lists allows duplicates, ordered and changeable data the lists in pyhton are preferable.
    # Since it follows sequence in which the data is received, the pyhton lists are recommended option. 
    
    inputData   = [] # input data file with list.
    outputData  = [] # output data file with list.
    
    # Why do we need training Data? : 
    # The traning data set is used to train the model of the learning algorithm to fit the targets.
    # Say we train for House Prices Based on history of the land, current prices,sq. ft., rooms etc.
    # We train the model for different available houses with the above parameters to improve the accuracy of model.
    # Here the Training Data is Available Houses with given parameters to train the model.
    # If there is no training data the algorithm won't be able to predict the future prices without their analysis.
    
    for i in range(start,end): #defining a range using for loop for the input data. 
        inputData.append(i) #add the varibale value at the end of the input file.
        outputData.append(fizzbuzz(i)) #add/modify the varibale value at the end of the output file.
    
    # Why Dataframe? : 
    # The data frames are used to store the values in the form of grids which are easy to look upon.
    # A Dataframe generally contains data, index & columns which can be accessed easily and are simple to manage.
    
    dataset = {} #data in the frame.
    dataset["input"]  = inputData #index as input in the frame.
    dataset["label"] = outputData #colums as labels in the frame. 
    
    # Writing to csv 
    pd.DataFrame(dataset).to_csv(filename) #The frame is written to a csv file with the above written parameters.
    
    print(filename, "Created!")
```

## Processing Input and Label Data

```python
def processData(dataset): #A function processData is defined with input from data
    
    # Why do we have to process? :
    # The data value we get from the dataset needs a certain process to be encoded.
    # But before that they have to be cleaned of the gliches which can occur during data tranfer from one file to another file.
    # Thus we create a processLabel to record the data which we have to encode later on with a defined process to ensure the encoding of the data without errors.
    
    
    data   = dataset['input'].values #from dataset access/import the values of input. 
    labels = dataset['label'].values #from labels access/import the values of labels.
    
    processedData  = encodeData(data) #get the encoded data in the process data.
    processedLabel = encodeLabel(labels) #get the encoded lable in the process lable.
    
    return processedData, processedLabel #return the value as processData & processLable with encoded values.
```

```python
def encodeData(data): #A function encodeData is defined with input from data
    
    processedData = [] #processedData function called with its values. 
    
    for dataInstance in data: #Initialize for loop to check the dataInstance in data.
        
        # Why do we have number 10? :
        # We have our training set data from 101-1001 we have to select a range that wont affect the process.
        # Since we have to append the dataInstances from the data inputs we use range as 10.
        # Range in 0-9 so that the incoming data is managable and does not affected the append process.
        
        processedData.append([dataInstance >> d & 1 for d in range(10)]) 
        #add the acquired data from dataInstance to processedData. 
    
    return np.array(processedData) #add and create an array of the acquired values of processedData. 
```

```python
from keras.utils import np_utils #import numpy utils from keras.utils to transform data into numerical sequences.

def encodeLabel(labels): #A function encodeLabels is defined with input from labels.
    
    processedLabel = [] #processedLabel function is called with its values
    
    for labelInstance in labels: #Initialize for loop to check the dataInstance in labels.
        if(labelInstance == "FizzBuzz"):
            # Fizzbuzz
            processedLabel.append([3]) #labelInstacnce = FizzBuzz from encodedLabel, append FizzBuzz in processedLabel.
        elif(labelInstance == "Fizz"):
            # Fizz
            processedLabel.append([1]) #labelInstacnce = Fizz from encodedLabel,, append Fizz in processedLabel.
        elif(labelInstance == "Buzz"):
            # Buzz
            processedLabel.append([2]) #labelInstacnce = Buzz from encodedLabel, append Buzz in processedLabel.
        else:
            # Other
            processedLabel.append([0]) #else from encodedLabel, append other value in processedLabel.

    return np_utils.to_categorical(np.array(processedLabel),4) 
    #return the values from categorical array labels with 4 values 0,1,2,3 according to the for loop comparison.
```

## Model Definition

```python
from keras.models import Sequential #import the sequential analysis model from keras.
from keras.layers import Dense, Activation, Dropout #import layered models from keras.
from keras.callbacks import EarlyStopping, TensorBoard #import callback functions from kera.
from keras.utils import np_utils #import numpy utils from keras.utils
import numpy as np #import numpy library as np 

input_size = 10 #input size selected as 10.
drop_out = 0.2 #drop out / ignore rate is set to 0.2.
first_dense_layer_nodes  = 256 #first layer nodes for network activation and processing. 
second_dense_layer_nodes = 4 #second layer nodes for deeper network activation and processing.

def get_model():
    
    # Why do we need a model? : 
    # A model is nothing but a set of specific instructions which are defined.
    # And every process has to utilize the same model for its execution.
    # Say we want to boil eggs, the steps we take to do so are 
    # 1-Get required number of eggs ; 2-Get a sizeable pan in which the eggs will fit ; 3- Put the eggs in the pan
    # 4-Keep the pan on the heating flame and let the water boil ; 5-Estimated time is around 15mins ;
    # 6-After sufficient time passes turn the flame off ; 7-Cool the eggs in another container with cold water
    # 8-Remove the Coverings ; 9-Cut the eggs and Eat with salt pepper if required.
    # Similarly with the process we follow a model for execution; this way we can manage, execute and analyse the process in detail.

    # Why use Dense layer and then activation? :
    # Using the dense layer first gives an insight in using actvations, also we can retrieve outputs of the last layer.
    # This is not possible in the mixed activation where both layers aregiven at the same time.
    
    # Why use sequential model with layers?
    # Sequential model allows layered structure, i.e. we can create layer by layer model.
    # This gives us a detailed idea of working of each layer and also evaluating the model becomes less difficult.
    # Also the modification and data fitting is comparatively easier. 
    
    model = Sequential()
    
    model.add(Dense(first_dense_layer_nodes, input_dim=input_size))
    model.add(Activation('relu'))
    
    # Why dropout? : 
    # Drop out is a simple learning regularization technique to prevent overfitting of data.
    # Also this includes ignoring of randomly selected neurons in a network.
    
    model.add(Dropout(drop_out))
    
    model.add(Dense(second_dense_layer_nodes))
    model.add(Activation('softmax'))
    
    # Why Softmax?
    # Softmax is a logistc regression which is used when there involves classification of multiple classes.
    # Since here we activate softmax as the second dense layer function it acts as the final classifier for the learning model. 
    
    model.summary()
    
    # Why use categorical_crossentropy?
    # Cross-entropy loss, measures the performance of a classification model whose output is a probability value between 0 and 1. 
    # Cross-entropy loss increases as the predicted probability diverges from the actual label.
    # Also it is suggested to use this loss function when targets are one hot encoded, which is how we have encoded our labels.
    # That's why we use catgorical_crossentropy

    
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    #compile model using three parameters i.e. optimizer, loss and metrics.
    
    return model
```

# <font color='blue'>Creating Training and Testing Datafiles</font>

```python
# Create datafiles
createInputCSV(101,1001,'training.csv') # create the datafile training.csv for training set 101-1000
createInputCSV(1,101,'testing.csv') # create the datafile testing.csv for testing set 1-101
```

# <font color='blue'>Creating Model</font>

```python
model = get_model() # get the model in grid pattern from the defined parameters.
```

# <font color = blue>Run Model</font>

```python
validation_data_split = 0.2  # sets the data split for training and validation 
num_epochs = 10000  # sets the epochs 
model_batch_size = 128 
tb_batch_size = 32
early_patience = 100  # number of epochs to wait before early stop. 

tensorboard_cb   = TensorBoard(log_dir='logs', batch_size= tb_batch_size, write_graph= True) # tensorboard callback function.
earlystopping_cb = EarlyStopping(monitor='val_loss', verbose=1, patience=early_patience, mode='min')# early stopping function parmaeters.

# Read Dataset : Read the dataset training.csv
dataset = pd.read_csv('training.csv')

# Process Dataset : Process the dataset with the defined parameters.
processedData, processedLabel = processData(dataset)
history = model.fit(processedData
                    , processedLabel
                    , validation_split=validation_data_split
                    , epochs=num_epochs
                    , batch_size=model_batch_size
                    , callbacks = [tensorboard_cb,earlystopping_cb]
                   )
```

# <font color = blue>Training and Validation Graphs</font>

```python
%matplotlib inline 
df = pd.DataFrame(history.history) # pickup history from the history dataframe. 
df.plot(subplots=True, grid=True, figsize=(10,15)) # plot the graphs of the subplots
```

# <font color = blue>Testing Accuracy [Software 2.0]</font>

```python
def decodeLabel(encodedLabel): # define decodeLabel to fetch data from encodedLabel
    if encodedLabel == 0: 
        return "Other"
    elif encodedLabel == 1:
        return "Fizz"
    elif encodedLabel == 2:
        return "Buzz"
    elif encodedLabel == 3:
        return "FizzBuzz"
```

```python
wrong   = 0 
right   = 0

testData = pd.read_csv('testing.csv') # read the testing.csv file.

processedTestData  = encodeData(testData['input'].values) # encoded test data with input values in processedTestData function.
processedTestLabel = encodeLabel(testData['label'].values) # encoded test data with label values in processedTestLabel function.
predictedTestLabel = []

for i,j in zip(processedTestData,processedTestLabel): # zip combines two iterators in one for loop.
    y = model.predict(np.array(i).reshape(-1,10)) # generates output predictions for the input samples.
    predictedTestLabel.append(decodeLabel(y.argmax())) # append the newly predicted value in predictedTestLabel
    
    if j.argmax() == y.argmax(): # comparing of two arrays  
        right = right + 1 
    else:
        wrong = wrong + 1

print("Errors: " + str(wrong), " Correct :" + str(right)) # print number of errors and correct answers.

print("Testing Accuracy: " + str(right/(right+wrong)*100)) # print testing accuracy.

# Please input your UBID and personNumber 
testDataInput = testData['input'].tolist() 
testDataLabel = testData['label'].tolist()

testDataInput.insert(0, "UBID") #UBID inserted into output.csv file
testDataLabel.insert(0, "nabhyank")

testDataInput.insert(1, "personNumber")#UB-person number inserted into output.csv file
testDataLabel.insert(1, "50290958")

predictedTestLabel.insert(0, "")
predictedTestLabel.insert(1, "")

output = {}
output["input"] = testDataInput
output["label"] = testDataLabel

output["predicted_label"] = predictedTestLabel #compare the predictedTestLables with predicted_labels.

opdf = pd.DataFrame(output) #print the data frame into output file.
opdf.to_csv('output.csv') #convert the output file to .csv file.
```
